```python
import random
from graphviz import Digraph
from IPython.display import Image
```

```python
# Function to generate a random graph
def generate_random_graph(num_nodes, edge_probability):
    adj_list = {}                                                        #create an empty dictionary.
    # adj_list[1]=[3]
    # print(adj_list)
    for node in range(1, num_nodes + 1):                                 # creates new nodes like 1,2,3,4...
         adj_list[node] = []
     # print(adj_list[node])
     # print(adj_list.keys())

    for node in range(1, num_nodes + 1):
        for neighbor in range(1, num_nodes + 1):                         # to iterate the edge values.
            if node != neighbor and  random.random() < edge_probability: #checks no nodes have edge to itself and random<prob is used to limit the edges
                adj_list[node].append(neighbor)                          #adds edges to the current node.
     # print(adj_list)
    # print(adj_list[2])
    
    return adj_list
```

```python
 print(random.random())
```

```python
# Example parameters
num_nodes = 5
edge_probability = 0.5
```

```python
# Generate a random graph
adj_list=generate_random_graph(num_nodes, edge_probability)
print(adj_list)
```

```python
# Create a new graph
g = Digraph()
```

```python
# Add nodes to the graph
for node in adj_list:                          #this will iterate over the keys in adj_list dict.
    g.node(str(node))                          #adds node to the graph.

```

```python
# Add edges to the graph
for node, neighbors in adj_list.items():        #takes the node and edge values from the dictionary.
    # print(node,neighbors)                       
    for neighbor in neighbors:                  #iterate over the edge values in each node
        g.edge(str(node), str(neighbor) )       #this will create an edge between the node values between node and neighbor
        # print("hi")
```

```python
# from IPython.display import Image

g.render('random_graph', format='png', cleanup=True)
Image('random_graph.png')
```
