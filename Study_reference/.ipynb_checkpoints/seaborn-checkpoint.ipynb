```python
from sklearn.datasets import load_breast_cancer
breast_cancer=load_breast_cancer()
```

```python
breast_cancer
```

```python
cancer = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
cancer
```

```python
cancer['alive']= breast_cancer.target
cancer
```

```python
cancer=cancer.drop(columns='result', axis=1)
```

```python
cancer
```

```python
cancer.head()
```

```python
sns.distplot(cancer['mean radius'])
```

```python
# correlation matrix
# positive and negative:
correlation= cancer.corr()
```

```python
correlation
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8} )
```
