```python
import sklearn
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data = np.genfromtxt(r'E:\Demo Datasets\test_bird.csv', delimiter=',')
```

```python
X = data[1:,8]
X
```

```python
X =X.reshape(-1,1)
X.shape
```

```python
Y= data[1:,9]
Y
```

```python
Y = Y.reshape(-1,1)
Y.shape
```

```python
model = linear_model.LinearRegression(fit_intercept=True)
```

```python
model.fit(X,Y)
print(model.coef_)
print(model.intercept_)
```

```python
data_test = np.genfromtxt(r"E:\Demo Datasets\train_bird.csv", delimiter=',')
```

```python
X_test = data_test[1:,8]
Y_test = data_test[1:,9]
X_test = X_test.reshape((-1,1))
Y_test = Y_test.reshape((-1,1))
print(Y_test.shape)
Y_pred = model.predict(X_test)
```

```python
error = mean_squared_error(Y_test,Y_pred)
```
